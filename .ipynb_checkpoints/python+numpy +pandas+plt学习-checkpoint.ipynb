### *设置图片大小，和清晰程度

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
fig = plt.figure(figsize=(20,8),dpi=80)    #设置图片大小，和清晰程度
plt.plot(x,y)
```

### 设置坐标抽刻度---设置步长+设置字符串坐标轴

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
plt.xticks(range(2,24,0.5))   
plt.title("set start ticks")
plt.plot(x,y)
# 设置步长为0.5会出错，不允许float类型数据出现
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
plt.xticks(range(2,24,1))   
plt.title("设置步长为1")
plt.plot(x,y)
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
_x_labels = [i/2 for i in range(4,50)]
plt.figure(figsize = (20,8),dpi = 80)
plt.xticks(_x_labels)   
plt.title("set start ticks")
plt.plot(x,y)
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
plt.xticks(x[::2])   #也可以这样取步长
plt.title("set start ticks")
plt.plot(x,y)
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
plt.xticks(range(2,24,1)) 
plt.yticks(range(min(y),max(y)+1))
plt.title("set start ticks")
plt.plot(x,y)
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
_x = x
_x_xticks_labels = ["hello,{}".format(i) for i in _x]
plt.figure(figsize = (10,8),dpi = 80)
plt.xticks(_x,_x_xticks_labels)   
plt.title("set start ticks")
plt.plot(x,y)
```

```python
x = range(2,26,2)
y = [15,13,14,17,20,26,26,24,22,18,15,19]
plt.xticks(x[::2])   #也可以这样取步长
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("测试数据")
plt.plot(x,y)
```

### 添加网格化 方便阅读

```python
x = range(2,26,2)
y1 = [15,13,14,17,20,26,26,24,22,18,15,19]
y2 = [i+1 for i in y1]
plt.xticks(x[::2])   #也可以这样取步长
plt.title("set start ticks")
plt.grid(alpha = 0.4)  # 网格设置，并且设置透明度
plt.plot(x,y1)
plt.plot(x,y2)
```

### 双折线做图---并添加图例

```python
x = range(2,26,2)
y1 = [15,13,14,17,20,26,26,24,22,18,15,19]
y2 = [i+1 for i in y1]
plt.figure(figsize = (20,8),dpi = 80)
plt.plot(x,y1,x,y2)
plt.legend(["my","you"] ,loc = 1)
plt.title("set start ticks")
plt.grid(alpha = 0.4)  # 网格设置，并且设置透明度
```

```python
x = np.arange(0,10,step=1)
plt.plot(x,x,x,x*2,x,x/2)# 传入一个列表参数，参数内容就是每一个data的说明
plt.legend(['fast','normal','slow'])
```

### plt布局

```python
x = np.arange(0,10,step=1)
y = [2*i for i in x]
z = [3*i for i in x]
```

```python
import matplotlib.pyplot as plt
 
fig1, (ax1,ax2) = plt.subplots(ncols=2, nrows=1, constrained_layout=True,figsize = (10,4))

x = np.arange(0,10,step=1)

ax1.plot(x,y)
ax2.plot(x,z)

plt.show()
```
