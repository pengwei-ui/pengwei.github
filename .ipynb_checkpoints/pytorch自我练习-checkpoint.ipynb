### 构件模型--Linear(a,b)  输入有几个变量，输出有几个变量

```python
# class LinearRegresssion(nn.Module):
#     def __init__(self):
#         super(LinearRegresssion,self).__init__()
#         self.fc = nn.Linear(1,1)    
        
#     def forward(self,x):
#         out = self.fc(x)
#         return out
```

#### 定义模型

```python
loss_fn = nn.MSELoss()
```

#### 优化器--SGD 随机梯度下降

```python
opt = torch.optim.SGD(model.parameters(),lr = 0.001)  
```

```python
for name ,parameters in model.named_parameters():
    print("name :",name,"parame :",parameters )
```

```python
a ,b = [],[]
for epoch in range(100):
    for x,y in zip(X,Y):
        y_pred = model(x)        # 使用模型预测
        loss = loss_fn(y,y_pred) # 进行预测结果计算损失
        opt.zero_grad()          # 把梯度清0
        loss.backward()          # 求解梯度
        opt.step()               # 优化模型参数    
    with torch.no_grad():
        a.append(epoch)
        b.append(loss_fn(y,y_pred).data.numpy())
plt.scatter(a,b)
```

```python
from torch import nn
import torch.nn.functional as F 
```

#### 设置层级---我设置了1个中间层--中间层变量为3

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.liner_1 = nn.Linear(3,3)
        self.liner_2 = nn.Linear(3,1)
        
    def forward(self,input):
        x = F.relu(self.liner_1(input))
        x = F.relu(self.liner_2(x))
        return x
```

```python
model = Model()
model
```

#### 使用Adam进行替代SGD

```python
lr = 0.006
```

```python
def get_model():
    model = Model()
    opt = torch.optim.Adam(model.parameters(), lr =lr ,betas=(0.9,0.99))
    return model,opt
```

```python
model,optim = get_model()
```

#### 定义损失函数

当时二分类问题是使用BCELoss

当时解决线性关系时使用MSELoss

```python
loss_fn = nn.MSELoss()
```

#### 开始拟合

```python
a,b = [],[]
epochs = 100
for epoch in range(epochs):
    for x,y in zip(X,Y):
        y_pred = model(x)
        loss = loss_fn(y_pred,y)
        optim.zero_grad()
        loss.backward()
        optim.step()
    with torch.no_grad():
        a.append(epoch)
        b.append(loss_fn(model(X),Y).data.item())
plt.scatter(a,b)
plt.title("loss = {}, lr ={}".format(b[epochs-1],lr))
```

```python
y_now = model(test_X)
loss = loss_fn(y_now,test_Y)
loss.data.numpy() 
```

#### 这是使用SGD梯度下降算法的拟合结果

```python
class SGD_Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.liner_1 = nn.Linear(3,3)
        self.liner_2 = nn.Linear(3,1)
        
    def forward(self,input):
        x = F.relu(self.liner_1(input))
        x = F.relu(self.liner_2(x))
        return x
```

#### 这是使用RMSprop 下降法进行拟合

## 多分类问题--使用softmax

与二分类的区别，分类数量大于二，并且最后一层不在是sigmoid层（sigmoid结果只能是0,1）

损失函数：CrossEntropyLoss(里面直接就包含了softmax函数，故不用在自己设置的模型里添加softmax函数进行激活，使用relu和线性函数即可)

问题一: Y的取值范围必须是从0开始，不能从一开始，比如本题原始数据是从1到3三类数据，error:Target 3 is out of bounds.，这时应该修改原始数据，让其从0开始.或者在计算loss时让y-1

问题二：输入的label(Y)必须是长整型的Tensor,而不再是float数据类型，书写：train_y =torch.from_numpy(y_train).type(torch.LongTensor)

问题三： 当设置变量y时大小必须为【m】而不能是【m,1】因为计算loss_fn = loss(y_pred,y) 会报错 0D or 1D target tensor expected, multi-target not supported

#### 划分数据集和训练集

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_data,y_data)
train_x = torch.from_numpy(x_train).type(torch.float32)
train_y = torch.from_numpy(y_train).type(torch.LongTensor)
test_x = torch.from_numpy(x_test).type(torch.float32)
test_y = torch.from_numpy(y_test).type(torch.LongTensor)
train_x.size()
```

#### 随机切分训练集

```python
from torch.utils.data import TensorDataset,DataLoader
```

```python
xy_data = TensorDataset(train_x,train_y)
train_xy_data = DataLoader(xy_data,batch_size= 500,shuffle= True)
train_xy_data
```

#### 构建模型

```python
import torch.nn.functional as F
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(21,21)
        self.linear2 = nn.Linear(21,3)
        
    def forward(self,input):
        x = F.relu(self.linear1(input))
        output = F.relu(self.linear2(x))
        return output 
```

```python
model = Model()
model
```

#### 构造损失函数

```python
loss_fn = nn.CrossEntropyLoss()
input_batch,label_batch  = next(iter(train_xy_data))
input_batch.shape,label_batch.shape
```

重点： 这个与之前的有所不同，这个使用CrossEntropyLoss时的y变量维度不是1，可以看成是0

#### 构造优化器

```python
lr = 0.009
epochs =100
```

```python
opt = torch.optim.Adam(model.parameters(),lr =lr , betas= (0.9,0.99))
```

#### 确定返回值并且计算正确率

```python
def accuracy(y_pred,y_true):
    y_pred = torch.argmax(y_pred,dim =1)    # 查找Y返回的值
    acc = (y_pred==y_true-1).float().mean()
    return acc
```

#### 开始训练

```python
train_loss = []
train_acc = []
test_loss = []
test_acc = []
for epoch in range(epochs):
    for x,y in train_xy_data:
        y_pred = model(x)
        loss = loss_fn(y_pred,y-1)
        opt.zero_grad()
        loss.backward()
        opt.step()
    with torch.no_grad():
        epoch_acc = accuracy(model(train_x),train_y)
        epoch_loss = loss_fn(model(train_x),train_y-1).data
        
        epoch_test_acc = accuracy(model(test_x),test_y)
        epoch_test_loss = loss_fn(model(test_x),test_y-1).data
         
        train_loss.append(epoch_loss.item())
        train_acc.append(epoch_acc.item())
        test_loss.append(epoch_test_loss.item())
        test_acc.append(epoch_test_acc.item())


fig1, (ax1,ax2) = plt.subplots(ncols=2, nrows=1, constrained_layout=True,figsize = (10,4))

ax1.plot(range(1,epochs+1),train_loss,label = 'train_loss')
ax1.plot(range(1,epochs+1),test_loss,label = 'test_loss')
ax1.legend()
ax1.set_title("损失率")

ax2.plot(range(1,epochs+1),train_acc,label = 'train_acc')
ax2.plot(range(1,epochs+1),test_acc,label = 'test_acc')
ax2.legend()
ax2.set_title("正确率")

fig1.show()
```
